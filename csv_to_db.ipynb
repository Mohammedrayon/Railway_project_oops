```python
import pandas as pd
import numpy as np

df = pd.read_csv('train_info.csv')
df
```

```python
# store the serial no ,Train_No , Train_Name , Source_Station , Destination_Station in sqllite database
import sqlite3
conn = sqlite3.connect('train.db')
c = conn.cursor()
c.execute('CREATE TABLE train_info (serial_no int, Train_No int, Train_Name text, Source_Station text, Destination_Station text)')
conn.commit()
df.to_sql('train_info', conn, if_exists='replace', index = False)
c.execute('''
SELECT * FROM train_info
          ''')
for row in c.fetchall():
    print(row)
conn.commit()
conn.close()
```
