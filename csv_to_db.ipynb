```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('database/train.db')
c = conn.cursor()

# Create the table
c.execute('''
CREATE TABLE IF NOT EXISTS train_info (
    Train_No INTEGER PRIMARY KEY,
    Train_Name TEXT,
    Source_Station TEXT,
    Destination_Station TEXT
)
''')
conn.commit()

# Select only the required columns from the DataFrame
df = df[['Train_No', 'Train_Name', 'Source_Station_Name', 'Destination_Station_Name']]

# Rename the columns to match the database schema
df.columns = ['Train_No', 'Train_Name', 'Source_Station', 'Destination_Station']

# Store the DataFrame in the SQLite database
df.to_sql('train_info', conn, if_exists='replace', index=False)

# Query to select all data from the table
c.execute('SELECT * FROM train_info')

# Fetch and print all rows
for row in c.fetchall():
    print(row)

conn.close()
```

```python
import random 

def generate_pnr():
    region_code = random.choice(['1', '2', '3', '4', '5', '6', '7', '8', '9'])  # First digit
    remaining_digits = ''.join([str(random.randint(0, 9)) for _ in range(9)])  # Remaining 9 digits
    pnr_number = region_code + remaining_digits
    return pnr_number

print(generate_pnr())
```
